```python
import nltk
from nltk.tree import Tree
```

```python
# Define the three parse trees

# 1. "The government raised interest rates."
tree1 = Tree('S', [
    Tree('NP', [
        Tree('Det', ['The']),
        Tree('N', ['government'])
    ]),
    Tree('VP', [
        Tree('V', ['raised']),
        Tree('NP', [
            Tree('Det', ['interest']),
            Tree('N', ['rates'])
        ])
    ])
])

# 2. "The internet gives everyone a voice."
tree2 = Tree('S', [
    Tree('NP', [
        Tree('Det', ['The']),
        Tree('N', ['internet'])
    ]),
    Tree('VP', [
        Tree('V', ['gives']),
        Tree('NP', [
            Tree('Pro', ['everyone'])
        ]),
        Tree('NP', [
            Tree('Det', ['a']),
            Tree('N', ['voice'])
        ])
    ])
])

# 3a "The man saw the dog with the telescope."
# meaning "with the telescope the man saw the dog"
tree3_option1 = Tree('S', [
    Tree('NP', [
        Tree('Det', ['The']),
        Tree('N', ['man'])
    ]),
    Tree('VP', [
        Tree('V', ['saw']),
        Tree('NP', [
            Tree('Det', ['the']),
            Tree('N', ['dog'])
        ]),
        Tree('PP', [
            Tree('P', ['with']),
            Tree('NP', [
                Tree('Det', ['the']),
                Tree('N', ['telescope'])
            ])
        ])
    ])
])

# 3b "The man saw the dog with the telescope."
# meaning "the dog with the telescope was seen by the man"
tree3_option2 = Tree('S', [
    Tree('NP', [
        Tree('Det', ['The']),
        Tree('N', ['man'])
    ]),
    Tree('VP', [
        Tree('V', ['saw']),
        Tree('NP', [
            Tree('Det', ['the']),
            Tree('N', ['dog']),
            Tree('PP', [
                Tree('P', ['with']),
                Tree('NP', [
                    Tree('Det', ['the']),
                    Tree('N', ['telescope'])
                ])
            ])
        ])
    ])
])

# Display the trees
print("\nTree 1: 'The government raised interest rates.'")
#tree1.pretty_print()

print("\nTree 2: 'The internet gives everyone a voice.'")
#tree2.pretty_print()

print("\nTree 3 (Option 1): 'The man saw the dog with the telescope.' (telescope modifies saw)")
#tree3_option1.pretty_print()

print("\nTree 3 (Option 2): 'The man saw the dog with the telescope.' (telescope modifies dog)")
#tree3_option2.pretty_print()
```
